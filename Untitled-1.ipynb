```python
# 1. Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# 2. Basic statistics for numerical columns
print("\nDescriptive statistics for numerical columns:")
print(df.describe(include='number'))

# 3. Churn distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Churn', data=df)
plt.title('Distribution of Churn')
plt.show()

# 4. Churn rate by categorical features
categorical_cols = df.columns[df.dtypes == 'object'].tolist()

plt.figure(figsize=(20, 30))
for i, col in enumerate(categorical_cols, 1):
    plt.subplot(6, 3, i)
    sns.barplot(x=col, y='Churn', data=df, estimator=lambda x: len(x[x=='Yes'])/len(x)*100)
    plt.title(f'Churn Rate by {col}')
    plt.ylabel('Churn Rate (%)')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 5. Numerical features vs churn
numerical_cols = df.columns[df.dtypes != 'object'].tolist()

plt.figure(figsize=(20, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(1, 10, i)
    sns.boxplot(x='Churn', y=col, data=df)
    plt.title(f'{col} vs Churn')
plt.tight_layout()
plt.show()

# 6. Correlation matrix
# Convert churn to numerical for correlation
#df['churn_numeric'] = df['churn'].apply(lambda x: 1 if x == 'Yes' else 0)

plt.figure(figsize=(10, 8))
corr = df[numerical_cols ].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

```python
df.drop(columns=df.select_dtypes(include='object'), inplace=True)# Drop fcategorical categorical column for simplicity
#  Split data into features and target
x= df.drop(columns=['Churn'])
y= df['Churn']
#  Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)
#  Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 7. Check class imbalance
print("\nClass distribution:")
print(df['Churn'].value_counts(normalize=True))

# 10. Handle class imbalance with SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)

print("\nData shapes after preprocessing:")
print(f"Original train shape: {X_train.shape}")
print(f"After SMOTE: {X_train_smote.shape}")
print(f"Test shape: {X_test.shape}")
print(f"after Smote: {y_train_smote.value_counts()}")
```

```python
df.columns,x.columns
```

```python
# 1. Logistic Regression (baseline model)
print("\nLogistic Regression Results:")
lr = LogisticRegression(random_state=42, max_iter=1000)
lr.fit(X_train_smote, y_train_smote)
y_pred_lr = lr.predict(X_test_scaled)
y_prob_lr = lr.predict_proba(X_test_scaled)[:, 1]

print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_lr))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_lr))
print("\nROC AUC Score:", roc_auc_score(y_test, y_prob_lr))

# 2. Random Forest Classifier
print("\nRandom Forest Results:")
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_smote, y_train_smote)
y_pred_rf = rf.predict(X_test_scaled)
y_prob_rf = rf.predict_proba(X_test_scaled)[:, 1]

print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rf))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))
print("\nROC AUC Score:", roc_auc_score(y_test, y_prob_rf))

# 3. Feature Importance from Random Forest
feature_importance = pd.DataFrame({
    'Feature': x.columns,
    'Importance': rf.feature_importances_
}).sort_values('Importance', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance)
plt.title(' Important Features for Churn Prediction')
plt.tight_layout()
plt.show()
```

```python
# Hyperparameter tuning for Random Forest
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, 
                          cv=5, n_jobs=-1, verbose=1, scoring='roc_auc')
grid_search.fit(X_train_smote, y_train_smote)

# Best parameters
print("\nBest parameters found:")
print(grid_search.best_params_)

# Best model
best_rf = grid_search.best_estimator_
y_pred_best = best_rf.predict(X_test_scaled)
y_prob_best = best_rf.predict_proba(X_test_scaled)[:, 1]

print("\nOptimized Random Forest Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_best))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_best))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_best))
print("\nROC AUC Score:", roc_auc_score(y_test, y_prob_best))
```

```python
# 1. Key drivers of churn from feature importance
top_features = feature_importance
print("\nTop drivers of churn:")
print(top_features)

# 2. Customer segments with highest churn risk
# Create a function to calculate churn probability
df['churn_probability'] = best_rf.predict_proba(x)[:, 1]

# Analyze high-risk segments
high_risk = df[df['churn_probability'] > 0.7]
print(f"\nNumber of high-risk customers: {len(high_risk)}")

# Characteristics of high-risk customers
print("\nCharacteristics of high-risk customers:")
print("Average tenure:", high_risk['Years'].mean())
print("Average monthly charges:", high_risk['Total_Purchase'].mean())
print("Contract types:")
print(high_risk['Years'].value_counts())
print(high_risk['Age'].value_counts())

# # 3. Calculate potential revenue loss
# avg_revenue = df['Total_Purchase'].mean()
# potential_loss = len(high_risk) * avg_monthly_revenue * 12  # Annual loss
# print(f"\nPotential annual revenue loss from high-risk customers: ${potential_loss:,.2f}")

# 4. Recommendations
print("\nBusiness Recommendations:")
print("1. Focus retention efforts on customers with month-to-month contracts")
print("2. Customers with fiber optic internet service have higher churn - investigate service quality")
print("3. High monthly charges correlate with churn - consider loyalty discounts")
print("4. Implement targeted outreach for customers identified as high-risk by the model")
print("5. Improve tech support services as lack of support is a significant churn driver")
```
